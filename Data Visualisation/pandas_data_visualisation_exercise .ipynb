# Pandas Data Visualization Exercise

This is just a quick exercise for you to review the various plots we showed earlier. Use **df3** .

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df3 = pd.read_csv('../data/df3')

%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

** Create this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion.**

```python
df3.plot.scatter(x='a', y='b', s=50, c='red', figsize=(12,3), edgecolor='black')
```

** Create a histogram of the 'a' column.**

```python
df3['a'].plot.hist(edgecolor='black', color='blue')
```

** These plots are okay, but they don't look very polished. Use style sheets to set the style to 'ggplot' and redo the histogram from above. Also figure out how to add more bins to it.***

```python
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(alpha=0.5, bins=20, edgecolor='white')
```

** Create a boxplot comparing the a and b columns.**

```python
df3.boxplot(column=['a','b'])
df3[['a','b']].plot.box()
```

** Create a kde plot of the 'd' column **

```python
df3['d'].plot.kde(xticks=np.arange(-0.5,1.6,step=0.5))
```

** Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line)**

```python
df3['d'].plot.kde(xticks=np.arange(-0.5,1.6,step=0.5), lw=4, ls='--')
```

** Create an area plot of all the columns for just the rows up to 30.**

```python
df3.loc[0:30].plot.area(stacked=False)
```

** Notice how the legend in our previous figure overlapped some of actual diagram. Can you figure out how to display the legend outside of the plot as shown below?**

```python
f = plt.figure()
df3.loc[0:30].plot.area(stacked=False, ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
